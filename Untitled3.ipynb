```python
import music21
from music21 import stream, note, midi
import random

# Create a stream for the melody
melody = stream.Stream()

# Generate 10 random notes (C4 to C5 range)
for _ in range(10):
    pitch = random.choice(['C4', 'D4', 'E4', 'F4', 'G4', 'A4', 'B4', 'C5'])
    n = note.Note(pitch)
    n.quarterLength = random.choice([0.5, 1, 1.5])  # Random durations
    melody.append(n)

# Save as MIDI
mf = midi.translate.streamToMidiFile(melody)
mf.open('random_melody.mid', 'wb')
mf.write()
mf.close()

# Play the melody (requires Colab's audio support)
from music21 import converter
s = converter.parse('random_melody.mid')
s.show('midi')  # This will attempt to play in Colab
```
